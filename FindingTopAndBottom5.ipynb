```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
CSP=pd.read_csv("data/CSP_ALL.csv")
```

```python
CSA=pd.read_csv("data/CSA_ALL.csv")
```

## CSP Top/Bottom 5

```python
CSP_B=CSP.groupby("State").mean()[" % Passed"].sort_values().head()
CSP_B_states=CSP_B.index
CSP_B
```

The states with the lowest average passing rate from 2017-2021 were Arkansas, District-of-Columbia, Louisiana, Mississippi, and Alabama.

```python
CSP_T=CSP.groupby("State").mean()[" % Passed"].sort_values(ascending=False).head()
CSP_T_states=CSP_T.index
CSP_T
```

The states with the highest average passing rate from 2017-2021 were Vermont, Nebraska, Wisconsin, Michigan, and Kansas.

## CSA Top/Bottom 5

```python
CSA_B=CSA.groupby("State").mean()[" % Passed"].sort_values().head()
CSA_B_states=CSA_B.index
CSA_B
```

The states with the lowest average passing rate from 2017-2021 were Arkansas, South Dakota, Wyoming, Montanna, and Mississippi.

```python
CSA_T=CSA.groupby("State").mean()[" % Passed"].sort_values(ascending=False).head()
CSA_T_states=CSA_T.index
CSA_T
```

The states with the highest average passing rate from 2017-2021 were Iowa, Michigan, New Hampshire, Kansas, and Alaska.

### CSP plots for Top and Bottom 5

```python
top_bottom_CSP=CSP[CSP["State"].isin(np.append(CSP_B_states, CSP_T_states))]



cond=[top_bottom_CSP["State"].isin(CSP_B_states), top_bottom_CSP["State"].isin(CSP_T_states)]
values=['Bottom', 'Top']
top_bottom_CSP['Divide']=np.select(cond,values)
```

```python
sns.scatterplot(data=top_bottom_CSP, x=" % Passed", y=" Total", hue = "State", style="Divide")
#plt.xscale('log')
plt.yscale('log')
plt.ylim([0,10000])
plt.xlim([0,100])
plt.title("Top/Bottom 5 CSP States for Passing Rates by Number of Exam Participants")
plt.ylabel("Total Participants")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
# fig=plt
# fig.tight_layout()
# plt.savefig("figures/CSP_color")
```

```python
sns.scatterplot(data=top_bottom_CSP, x=" % Passed", y=" Total", hue = "Divide")
plt.title("Top/Bottom 5 CSP States for Passing Rates by Number of Exam Participants")
plt.ylabel("Total Participants")
plt.yscale('log')
plt.ylim([0,10000])
plt.xlim([0,100])
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```
